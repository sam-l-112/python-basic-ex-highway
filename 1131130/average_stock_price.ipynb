```python
import twstock
stock = twstock.Stock('2331')
bfp = twstock.BestFourPoint(stock)
bfp.best_four_point_to_buy()   # 判斷是否為四大買點
bfp.best_four_point_to_sell()  # 判斷是否為四大賣點
bfp.best_four_point()          # 綜合判斷
```
